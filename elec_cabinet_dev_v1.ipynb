```python
class cabinet():
    def __init__(self, vol, vent_area, comb_area, hrrpua, tbur, tig, sim_no):
        self.volume = vol
        self.vent_area = vent_area
        self.comb_area = comb_area
        self.hrrpua    = hrrpua
        self.tbur      = tbur
        self.tig       = tig
        self.sim_no    = sim_no
        
    def head(self):
        l1 = '&HEAD CHID= \'cab_%d\', TITLE=\'Simulation No - %d\'/\n\n'%(self.sim_no,self.sim_no)
        l2 = '&TIME T_END=3000.00 /\n\n'
        return l1+l2

    def mesh(self):
        if self.volume == 0.01:
            l1 = '&MESH ID=\'MESH 01\' IJK= 14.00, 20.00, 40.00, XB=-0.35,0.35,-0.80,0.20,0.00,2.00 /\n\n'
            l2 = '&VENT MB=\'XMIN\', SURF_ID=\'OPEN\' /\n'
            l3 = '&VENT MB=\'XMAX\', SURF_ID=\'OPEN\' /\n'
            l4 = '&VENT MB=\'YMIN\', SURF_ID=\'OPEN\' /\n'
            l5 = '&VENT MB=\'YMAX\', SURF_ID=\'OPEN\' /\n'
            l6 = '&VENT MB=\'ZMAX\', SURF_ID=\'OPEN\' /\n\n'
            
        else:
            l1 = '&MESH ID=\'MESH 01\' IJK= 14.00, 20.00, 60.00, XB=-0.35,0.35,-0.80,0.20,0.00,3.00 /\n\n'
            l2 = '&VENT MB=\'XMIN\', SURF_ID=\'OPEN\' /\n'
            l3 = '&VENT MB=\'XMAX\', SURF_ID=\'OPEN\' /\n'
            l4 = '&VENT MB=\'YMIN\', SURF_ID=\'OPEN\' /\n'
            l5 = '&VENT MB=\'YMAX\', SURF_ID=\'OPEN\' /\n'
            l6 = '&VENT MB=\'ZMAX\', SURF_ID=\'OPEN\' /\n\n'
            
        return l1+l2+l3+l4+l5+l6
    
    def cab(self):
        if self.volume == 0.25:
            l1 = '&MATL ID = \'STEEL\', CONDUCTIVITY  = 54.00, SPECIFIC_HEAT = 0.47, DENSITY = 7833.00/\n'
            l2 = '&SURF ID = \'STEEL PLATE\', MATL_ID = \'STEEL\', THICKNESS     = 0.0028, COLOR = \'GRAY 50\'/\n\n'
            l3 = '&OBST XB = -0.25,0.25,-0.50,-0.50,0.00,1.00, SURF_ID=\'STEEL PLATE\'/ Front\n'
            l4 = '&OBST XB = -0.25,0.25,0.00,0.00,0.00,1.00, SURF_ID=\'STEEL PLATE\'/ Back\n'
            l5 = '&OBST XB = -0.25,-0.25,-0.50,0.00,0.00,1.00, SURF_ID=\'STEEL PLATE\'/ Left\n' 
            l6 = '&OBST XB = 0.25,0.25,-0.50,0.00,0.00,1.00, SURF_ID=\'STEEL PLATE\'/ Right\n'
            l7 = '&OBST XB = -0.25,0.25,-0.50,0.00,1.00,1.00, SURF_ID=\'STEEL PLATE\'/ Top\n'
            l8 = '&OBST XB = -0.25,0.25,-0.50,0.00,0.00,0.00, SURF_ID=\'STEEL PLATE\'/ Bottom\n\n'
            
        else:
            l1 = '&MATL ID = \'STEEL\', CONDUCTIVITY  = 54.00, SPECIFIC_HEAT = 0.47, DENSITY = 7833.00/\n'
            l2 = '&SURF ID = \'STEEL PLATE\', MATL_ID = \'STEEL\', THICKNESS     = 0.0028, COLOR = \'GRAY 50\'/\n\n'
            l3 = '&OBST XB = -0.25,0.25,-0.50,-0.50,0.00,2.00, SURF_ID=\'STEEL PLATE\'/ Front\n'
            l4 = '&OBST XB = -0.25,0.25,0.00,0.00,0.00,2.00, SURF_ID=\'STEEL PLATE\'/ Back\n'
            l5 = '&OBST XB = -0.25,-0.25,-0.50,0.00,0.00,2.00, SURF_ID=\'STEEL PLATE\'/ Left\n' 
            l6 = '&OBST XB = 0.25,0.25,-0.50,0.00,0.00,2.00, SURF_ID=\'STEEL PLATE\'/ Right\n'
            l7 = '&OBST XB = -0.25,0.25,-0.50,0.00,2.00,2.00, SURF_ID=\'STEEL PLATE\'/ Top\n'
            l8 = '&OBST XB = -0.25,0.25,-0.50,0.00,0.00,0.00, SURF_ID=\'STEEL PLATE\'/ Bottom\n\n'
            
        return l1+l2+l3+l4+l5+l6+l7+l8
    
    def ventilation_area(self):
        if self.volume == 0.25 and self.vent_area == 0.01:
            l1 = '&HOLE XB= -0.1,0.1,-0.5028,-0.4972,0.10,0.15/ Bottom Vent  Av = 0.01\n'
            l2 = '&HOLE XB= -0.1,0.1,-0.5028,-0.4972,0.85,0.90/ Top Vent     Av = 0.01\n\n'
        elif self.volume == 0.25 and self.vent_area == 0.02:
            l1 = '&HOLE XB= -0.2,0.2,-0.5028,-0.4972,0.10,0.15/ Bottom Vent  Av = 0.02\n'
            l2 = '&HOLE XB= -0.2,0.2,-0.5028,-0.4972,0.85,0.90/ Top Vent     Av = 0.02\n\n'
        elif self.volume == 0.5 and self.vent_area == 0.01:
            l1 = '&HOLE XB= -0.1,0.1,-0.5028,-0.4972,0.10,0.15/ Bottom Vent  Av = 0.01\n'
            l2 = '&HOLE XB= -0.1,0.1,-0.5028,-0.4972,1.85,1.90/ Top Vent     Av = 0.01\n\n'
        else:
            l1 = '&HOLE XB= -0.2,0.2,-0.5028,-0.4972,0.10,0.15/ Bottom Vent  Av = 0.02\n'
            l2 = '&HOLE XB= -0.2,0.2,-0.5028,-0.4972,1.85,1.90/ Top Vent     Av = 0.02\n\n'
        return l1+l2
    
    def burner(self):
        l1 = '&REAC FUEL = \'PROPANE\'/\n'
        l2 = '&SURF ID=\'BURNER\', HRRPUA = 1800, COLOR=\'RED\'/ HRRPUA=400 kW/m2, HRR = 20kW\n'
        l3 = '&VENT XB=-0.25,0.25,-0.10,-0.05,0.0,0.0, SURF_ID = \'BURNER\' /\n\n'
        return l1+l2+l3
    
    def mat_tig(self):
        if self.hrrpua == 75 and self.tig == 325:
            l1 = '&MATL ID=\'PVCWCFR\', DENSITY=1605., CONDUCTIVITY = 0.10, SPECIFIC_HEAT = 3.00 /\n'
            l2 = '&SURF ID=\'PVCWCFR\',COLOR=\'BLACK\',IGNITION_TEMPERATURE=325,HRRPUA=1,RAMP_Q=\'CONE\',CONE_HEAT_FLUX=65,MATL_ID=\'PVCWCFR\',THICKNESS=0.006/\n\n'
        elif self.hrrpua == 75 and self.tig == 400:
            l1 = '&MATL ID=\'PVCWCFR\', DENSITY=1605., CONDUCTIVITY = 0.10, SPECIFIC_HEAT = 3.00 /\n'
            l2 = '&SURF ID=\'PVCWCFR\',COLOR=\'BLACK\',IGNITION_TEMPERATURE=400,HRRPUA=1,RAMP_Q=\'CONE\',CONE_HEAT_FLUX=65,MATL_ID=\'PVCWCFR\',THICKNESS=0.006/\n\n'
        elif self.hrrpua == 150 and self.tig == 325:
            l1 = '&MATL ID=\'PVCWC\', DENSITY=1605., CONDUCTIVITY = 0.10, SPECIFIC_HEAT = 1.50 /\n'
            l2 = '&SURF ID=\'PVCWC\',COLOR=\'BLACK\',IGNITION_TEMPERATURE=325,HRRPUA=1,RAMP_Q=\'CONE\',CONE_HEAT_FLUX=65,MATL_ID=\'PVCWC\',THICKNESS=0.006/\n\n'
        elif self.hrrpua == 150 and self.tig == 400:
            l1 = '&MATL ID=\'PVCWC\', DENSITY=1605., CONDUCTIVITY = 0.10, SPECIFIC_HEAT = 1.50 /\n'
            l2 = '&SURF ID=\'PVCWC\',COLOR=\'BLACK\',IGNITION_TEMPERATURE=400,HRRPUA=1,RAMP_Q=\'CONE\',CONE_HEAT_FLUX=65,MATL_ID=\'PVCWC\',THICKNESS=0.006/\n\n'
        return l1+l2
    
    def combustible_area(self):
        if self.hrrpua == 75 and self.comb_area == 0.25:
            l1 = '&OBST XB=-0.25,0.25,-0.05,0.00,0.00,0.50, SURF_ID=\'PVCWCFR\' / combustible surface area = 0.25\n\n'
        elif self.hrrpua == 75 and self.comb_area == 0.50:
            l1 = '&OBST XB=-0.25,0.25,-0.05,0.00,0.00,1.00, SURF_ID=\'PVCWCFR\' / combustible surface area = 0.50\n\n'
        elif self.hrrpua == 150 and self.comb_area == 0.25:
            l1 = '&OBST XB=-0.25,0.25,-0.05,0.00,0.00,0.50, SURF_ID=\'PVCWC\' / combustible surface area = 0.25\n\n'
        else:
            l1 = '&OBST XB=-0.25,0.25,-0.05,0.00,0.00,1.00, SURF_ID=\'PVCWC\' / combustible surface area = 0.50\n\n'
        return l1
    
    def burning_duration(self):
        if self.tbur == 600:
            l1 = '&RAMP ID=\'CONE\', T=0.0, F=0.0/\n'
            l2 = '&RAMP ID=\'CONE\', T=100, F=%d/\n'%self.hrrpua
            l3 = '&RAMP ID=\'CONE\', T=200, F=%d/\n'%self.hrrpua
            l4 = '&RAMP ID=\'CONE\', T=500, F=%d/\n'%self.hrrpua
            l5 = '&RAMP ID=\'CONE\', T=600, F=0.0/\n\n'
        else:
            l1 = '&RAMP ID=\'CONE\', T=0.0, F=0.0/\n'
            l2 = '&RAMP ID=\'CONE\', T=100, F=%d/\n'%self.hrrpua
            l3 = '&RAMP ID=\'CONE\', T=200, F=%d/\n'%self.hrrpua
            l4 = '&RAMP ID=\'CONE\', T=600, F=%d/\n'%self.hrrpua
            l5 = '&RAMP ID=\'CONE\', T=800, F=%d/\n'%self.hrrpua
            l6 = '&RAMP ID=\'CONE\', T=900, F=0.0/\n\n'
        return l1+l2+l3+l4+l5
    
    def devices_hrr(self):
        l1 = '&DUMP DT_DEVC=2., SIG_FIGS=4 /\n'
        l2 = '&DEVC ID=\'q_bur\',QUANTITY=\'HRRPUA\', XYZ=0.00,-0.075,0.00, IOR=3 / Burner HRRPUA\n\n'
        if self.volume == 0.25:
            l3 = '&DEVC XB=-0.35,0.35,-0.80,0.20,0.00,2.00, QUANTITY=\'HRRPUV\', ID=\'HRR\', STATISTICS=\'VOLUME INTEGRAL\'/\n\n'
            l4 = '&DEVC ID=\'temp_top\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.00,-0.25,1.00, IOR=-3 / Material Temp at top of cabinet\n'
            l5 = '&DEVC ID=\'ghf_top\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.00,-0.25,1.00, IOR=-3 / Heat Flux at top of cabinet\n\n'
        else:
            l3 = '&DEVC XB=-0.35,0.35,-0.80,0.20,0.00,3.00, QUANTITY=\'HRRPUV\', ID=\'HRR\', STATISTICS=\'VOLUME INTEGRAL\'/\n\n'
            l4 = '&DEVC ID=\'temp_top\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.00,-0.25,2.00, IOR=-3 / Material Temp at top of cabinet\n'
            l5 = '&DEVC ID=\'ghf_top\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.00,-0.25,2.00, IOR=-3 / Heat Flux at top of cabinet\n\n'
        return l1+l2+l3+l4+l5
        
    def devices_mat(self):
        l1 = '&DEVC ID=\'temp_10\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.00,-0.05,0.10, IOR=-2 / Material Temp at 0.10 from base\n'
        l2 = '&DEVC ID=\'temp_20\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.00,-0.05,0.20, IOR=-2 / Material Temp at 0.20 from base\n'
        l3 = '&DEVC ID=\'temp_30\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.00,-0.05,0.30, IOR=-2 / Material Temp at 0.30 from base\n'
        l4 = '&DEVC ID=\'temp_40\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.00,-0.05,0.40, IOR=-2 / Material Temp at 0.40 from base\n'
        l5 = '&DEVC ID=\'temp_50\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.00,-0.05,0.48, IOR=-2 / Material Temp at 0.48 from base\n\n'

        l6 = '&DEVC ID=\'ghf_10\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.00,-0.05,0.10, IOR=-2 / Heat Flux at 0.10 from base\n'
        l7=  '&DEVC ID=\'ghf_20\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.00,-0.05,0.20, IOR=-2 / Heat Flux at 0.20 from base\n'
        l8 = '&DEVC ID=\'ghf_30\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.00,-0.05,0.30, IOR=-2 / Heat Flux at 0.30 from base\n'
        l9 = '&DEVC ID=\'ghf_40\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.00,-0.05,0.40, IOR=-2 / Heat Flux at 0.40 from base\n'
        l10= '&DEVC ID=\'ghf_50\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.00,-0.05,0.48, IOR=-2 / Heat Flux at 0.48 from base\n\n'
        return l1+l2+l3+l4+l5+l6+l7+l8+l9+l10
    
    def devices_mat_side(self):
        l1 = '&DEVC ID=\'temp_10_lm\', QUANTITY=\'WALL TEMPERATURE\', XYZ=-0.20,-0.05,0.10, IOR=-2 / Material Temp at 0.10 from base\n'
        l2 = '&DEVC ID=\'temp_20_lm\', QUANTITY=\'WALL TEMPERATURE\', XYZ=-0.20,-0.05,0.20, IOR=-2 / Material Temp at 0.20 from base\n'
        l3 = '&DEVC ID=\'temp_10_rm\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.20,-0.05,0.10, IOR=-2 / Material Temp at 0.10 from base\n'
        l4 = '&DEVC ID=\'temp_20_rm\', QUANTITY=\'WALL TEMPERATURE\', XYZ=0.20,-0.05,0.20, IOR=-2 / Material Temp at 0.20 from base\n\n'

        l5 = '&DEVC ID=\'ghf_10_lm\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=-0.20,-0.05,0.10, IOR=-2 / Heat Flux at 0.10 from base\n'
        l6 = '&DEVC ID=\'ghf_20_lm\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=-0.20,-0.05,0.20, IOR=-2 / Heat Flux at 0.20 from base\n'
        l7 = '&DEVC ID=\'ghf_10_rm\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.20,-0.05,0.10, IOR=-2 / Heat Flux at 0.10 from base\n'
        l8 = '&DEVC ID=\'ghf_20_rm\', QUANTITY=\'GAUGE HEAT FLUX\', XYZ=0.20,-0.05,0.20, IOR=-2 / Heat Flux at 0.20 from base\n\n'
        
        return l1+l2+l3+l4+l5+l6+l7+l8
    
    
    def devices_vent(self):
        if self.volume == 0.25:
            l1 = '&DEVC XB=0,0,-0.50,-0.50,0.885,0.885, ID=\'TC_topv\', QUANTITY=\'THERMOCOUPLE\' / Top Vent Temperature\n'
            l2 = '&DEVC XB=0,0,-0.50,-0.50,0.125,0.125, ID=\'TC_botv\', QUANTITY=\'THERMOCOUPLE\' / Bottom Vent Temperature\n'
            l3 = '&DEVC XB=0,0,-0.50,-0.50,0.885,0.885, ID=\'GHF_topv\', QUANTITY=\'GAUGE HEAT FLUX GAS\' / Top Vent gauge heatflux gas \n'
            l4 = '&DEVC XB=0,0,-0.50,-0.50,0.125,0.125, ID=\'GHF_botv\', QUANTITY=\'GAUGE HEAT FLUX GAS\' / Bottom Vent gauge heat flux gas\n\n'
        else:
            l1 = '&DEVC XB=0,0,-0.50,-0.50,1.875,1.875, ID=\'TC_topv\', QUANTITY=\'THERMOCOUPLE\' / Top Vent Temperature\n'
            l2 = '&DEVC XB=0,0,-0.50,-0.50,0.125,0.125, ID=\'TC_botv\', QUANTITY=\'THERMOCOUPLE\' / Bottom Vent Temperature\n'
            l3 = '&DEVC XB=0,0,-0.50,-0.50,1.875,1.875, ID=\'GHF_topv\', QUANTITY=\'GAUGE HEAT FLUX GAS\' / Top Vent gauge heatflux gas \n'
            l4 = '&DEVC XB=0,0,-0.50,-0.50,0.125,0.125, ID=\'GHF_botv\', QUANTITY=\'GAUGE HEAT FLUX GAS\' / Bottom Vent gauge heat flux gas\n\n'
        return l1+l2+l3+l4
    
    
    def bndf(self):
        l1 = '&BNDF QUANTITY=\'GAUGE HEAT FLUX\'  /\n'
        l2 = '&BNDF QUANTITY=\'BURNING RATE\'  /\n'
        l3 = '&BNDF QUANTITY=\'WALL TEMPERATURE\'  /\n'
        l4 = '&SLCF PBX=0.00 QUANTITY=\'TEMPERATURE\', VECTOR=.TRUE. /\n'
        return l1+l2+l3+l4
        
        
    def tail(self):
        return '&TAIL/\n\n'
    
    def writer(self):
        root_path = 'C:/Users/salviurvin/OneDrive - Virginia Tech/Desktop/docker/dec_07/'
        cab = 'cab_%d/'%self.sim_no
        fds_path = os.mkdir(os.path.join(root_path,cab))
        with open((root_path + 'cab_%d/cab.txt'%self.sim_no), 'w') as t:
            t.write(self.head())
            t.write(self.mesh())
            t.write(self.cab())
            t.write(self.ventilation_area())
            t.write(self.burner())
            t.write(self.mat_tig())
            t.write(self.combustible_area())
            t.write(self.burning_duration())
            t.write(self.devices_hrr())
            t.write(self.devices_mat())
            t.write(self.devices_mat_side())
            t.write(self.devices_vent())
            t.write(self.bndf())
            t.write(self.tail())
            
    
            
        

    

        
```

```python
def create_file(V, Av, Ac, hrrpua, tbur, tig, sim_no):
    a = cabinet(V, Av, Ac, hrrpua, tbur, tig, sim_no)
    a.writer()
    
def simulate(sim_no, root_path):
    root_path = root_path
    cab = 'cab_%d/'%sim_no
    os.system("start \"\" cmd /k \"fdsinit \
        \\& cd /D %s\
        \\& fds cab.txt"%(root_path+cab))
```
